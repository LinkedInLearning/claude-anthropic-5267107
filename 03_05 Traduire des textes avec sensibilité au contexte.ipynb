```python
ANTHROPIC_API_KEY = "XXXXXX"

client = anthropic.Anthropic(
    api_key=ANTHROPIC_API_KEY
)

def traduire_contexte(texte, langue_cible, tonalite="neutre", formel=True):
    prompt_systeme = (
        f"Tu es un traducteur expert. Traduis le texte suivant en {langue_cible}, "
        f"en respectant le contexte, le ton ({tonalite}) et le niveau de formalité ({'formel' if formel else 'informel'}). "
        f"N'adapte pas littéralement, mais conserve le sens et les intentions du texte original."
    )

    response = client.messages.create(
        model="claude-3-haiku-20240307",  # rapide, bon pour tâches linguistiques
        max_tokens=700,
        temperature=0.4,
        system=prompt_systeme,
        messages=[
            {"role": "user", "content": texte}
        ]
    )

    return response.content[0].text.strip()

# Exemple d'utilisation
texte_source = """Our team is excited to welcome new partners this year. Let's build something impactful together."""
langue = "français"
resultat = traduire_contexte(texte_source, langue_cible=langue, tonalite="enthousiaste", formel=False)

print("------------- Traduction contextuelle -------------\n")
print(resultat)
```
