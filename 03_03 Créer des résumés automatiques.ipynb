```python
import anthropic
import os
```

```python
ANTHROPIC_API_KEY = "XXXXXX"

client = anthropic.Anthropic(
    api_key=ANTHROPIC_API_KEY
)

def resume_texte(texte):
    system_prompt = "Tu es un assistant expert en résumé de texte. Résume le texte suivant de manière claire et concise."

    response = client.messages.create(
        model="claude-3-haiku-20240307",  # ou "claude-3-sonnet-20240229"
        max_tokens=300,
        temperature=0.5,
        system=system_prompt,
        messages=[
            {"role": "user", "content": texte}
        ]
    )

    return response.content[0].text.strip()

# Exemple de texte à résumer
texte_long = """
L'intelligence artificielle (IA) transforme de nombreux secteurs, allant de la santé à la finance. Elle permet d'automatiser des tâches répétitives, d'améliorer la prise de décision et d'accélérer la recherche. Cependant, elle soulève aussi des préoccupations éthiques, notamment en matière de biais, de transparence et de respect de la vie privée.
"""

# Résumé
resume = resume_texte(texte_long)
print("Résumé généré :\n", resume)
```

```python
import anthropic
import requests
from bs4 import BeautifulSoup

def extraire_texte_depuis_url(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Supprime les scripts/styles
    for tag in soup(["script", "style", "nav", "footer", "header"]):
        tag.decompose()

    # Récupère le texte brut
    texte = soup.get_text(separator=' ', strip=True)
    return texte

def resumer_texte(texte):
    prompt_systeme = "Tu es un assistant qui résume des articles web de façon claire, neutre et structurée."

    reponse = client.messages.create(
        model="claude-3-haiku-20240307",
        max_tokens=500,
        temperature=0.5,
        system=prompt_systeme,
        messages=[
            {"role": "user", "content": texte}
        ]
    )

    return reponse.content[0].text.strip()

url = "https://fr.wikipedia.org/wiki/Intelligence_artificielle"
texte = extraire_texte_depuis_url(url)
resume = resumer_texte(texte)

print("\n--------- Résumé de la page ---------n")
print(resume)
```
