# Execution via API

```python
# Étape 2 : Importer les bibliothèques nécessaires
import anthropic
```

```python
# Étape 3 : Définir votre clé API
client = anthropic.Anthropic(api_key="XXXXXX")
```

```python
# Étape 4 : Vérification
response = client.messages.create(
    model="claude-3-opus-20240229",  # ou "claude-3-sonnet-20240229", etc.
    max_tokens=300,
    messages=[
        {"role": "user", "content": "Explique-moi la différence entre l’IA faible et l’IA forte."}
    ]
)
print(response.content)
```
